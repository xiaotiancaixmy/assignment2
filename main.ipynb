# PyRamen

```python
import csv
from pathlib import Path

menu_file_path = Path('PyRamen', 'menu_data.csv')
sales_file_path = Path('PyRamen', 'sales_data.csv')

# Read the menu data
menu = []
with open(menu_file_path, 'r') as menu_file:
    menu_reader = csv.reader(menu_file)
    next(menu_reader)  
    for row in menu_reader:
        menu.append(row)

# Read the sales data
sales = []
with open(sales_file_path, 'r') as sales_file:
    sales_reader = csv.reader(sales_file)
    next(sales_reader)
    for row in sales_reader:
        sales.append(row)

# Initialize the report dictionary
report = {}
for row in sales:
    sales_item = row[4]
    quantity = int(row[3])

    # Check if sales_item is already included in the report dictionary
    if sales_item not in report:
        report[sales_item] = {
            '01-count': 0,
            '02-revenue': 0,
            '03-cogs': 0,
            '04-profit': 0
        }

    # Loop through the menu to find matching items
    for item in menu:
        if sales_item == item[0]:
            price = float(item[3])
            cost = float(item[4])
            profit = price - cost

            # Update the metrics in the report
            report[sales_item]['01-count'] += quantity
            report[sales_item]['02-revenue'] += price * quantity
            report[sales_item]['03-cogs'] += cost * quantity
            report[sales_item]['04-profit'] += profit * quantity
            break
    else:
        print(f"{sales_item} does not match with {item[0]}! Please double check the inventory")

# Write the report to a text file
report_file_path = Path('PyRamen', 'report.txt')
with open(report_file_path, 'w') as report_file:
    for item, metrics in report.items():
        report_file.write(f"{item} {metrics}\n")
```
