```python
# Set file paths for menu_data.csv and sales_data.csv
menu_file_path = "menu_data.csv"
sales_file_path = "sales_data.csv"

# Initialize list objects to hold menu and sales data
menu = []
sales = []

# Read in the menu data
with open(menu_file_path, 'r') as menu_file:
    next(menu_file)  # Skip the header row
    for line in menu_file:
        menu.append(line.strip().split(','))

# Read in the sales data
with open(sales_file_path, 'r') as sales_file:
    next(sales_file)  # Skip the header row
    for line in sales_file:
        sales.append(line.strip().split(','))

# Initialize dict object to hold key-value pairs of items and metrics
report = {}

# Initialize a row counter variable
row_count = 0

# Loop over every row in the sales data
for row in sales:
    # Extract the necessary columns from the sales data
    sales_item = row[4]

    # If the sales item is not in the report, add it as a new entry with initialized metrics
    if sales_item not in report:
        report[sales_item] = {
            "01-count": 0,
            "02-revenue": 0,
            "03-cogs": 0,
            "04-profit": 0
        }

    # Loop over the menu records to determine a match
    for menu_item in menu:
        if sales_item == menu_item[0]:
            # Extract the necessary columns from the menu data
            price = float(menu_item[3])
            cost = float(menu_item[4])

            # Calculate profit of each item in the menu data
            profit = price - cost

            # Update the metrics for the sales item in the report
            report[sales_item]["01-count"] += int(row[3])
            report[sales_item]["02-revenue"] += price * int(row[3])
            report[sales_item]["03-cogs"] += cost * int(row[3])
            report[sales_item]["04-profit"] += profit * int(row[3])

            # Print out matching menu data
            print(menu_item)

            # Break the loop as we found a match
            break
    else:
        # If no match is found, print the message
        print(f"{sales_item} does not equal any of the items in the menu data! No match!")

    # Increment the row counter by 1
    row_count += 1

# Print the total number of records in the sales data
print(f"Total number of records in sales data: {row_count}")

# Write out the report to a text file
with open("report.txt", "w") as report_file:
    for item, metrics in report.items():
        report_file.write(f"{item} {metrics}\n")
```
